```python
from diagrams import Cluster, Diagram
from diagrams.onprem.analytics import Spark
from diagrams.onprem.compute import Server
from diagrams.onprem.database import PostgreSQL
from diagrams.elastic.elasticsearch import Elasticsearch
from diagrams.onprem.inmemory import Redis
from diagrams.onprem.logging import Fluentd
from diagrams.onprem.monitoring import Grafana, Prometheus
from diagrams.onprem.network import Nginx
from diagrams.onprem.queue import Kafka, RabbitMQ
from diagrams.generic.device import Mobile, Tablet
from diagrams.generic.os import Windows


with Diagram("Enki",  direction="TB", show=True):
    with Cluster("Enki", direction="TB"):
        with Cluster("Backend"):
            backend = Server("Backend")
            rabbit = RabbitMQ("Async tasks")
            postgres= PostgreSQL("Postgres")
   
            keycloak = Server("Keycloak")

            backend >> rabbit
            backend >> postgres >> backend
            backend >> keycloak >> backend 
        
                
        with Cluster("FrontEnd"):
            
            fronts = [
                Mobile("Mobile"),
                Tablet("Tablette"), 
                Windows("Website")
            ]
            backend >> fronts
            fronts >> keycloak

        
    with Cluster("Nexsis"):
        with Cluster("SGE"):
            sge = Server("SGE")
            sge_pg = PostgreSQL("PG SGE")
            sge >> sge_pg >> sge
            kafka_sge = Kafka("Kafka")
            elastic_sge = Elasticsearch("Sge Event Hub")
            sge >> kafka_sge >> elastic_sge
        with Cluster("SGA",):
            sga = Server("SGA")
            sga_pg = PostgreSQL("PG SGA")
            sga >> sga_pg >> sga
        with Cluster("SGO"):
            sgo = Server("SGO")
            sgo_pg = PostgreSQL("PG SGO")
            sgo >> sgo_pg >>sgo
        with Cluster("SIG", direction="TB"):
            sig = Server("SIG")
            sig_pg = PostgreSQL("PG SIG")
            sig >> sig_pg >> sig

        sig >> sga 
        sig >> sgo
        sig >> backend
        sga >> sge >> sgo 
        
    kafka_sge >> backend
    backend >> elastic_sge
```
